```python
import time
import random # Using random for the demo, but swap for yfinance later
from collections import deque
from datetime import datetime

class MarketSentinel:
    def __init__(self, ticker, window_size=5):
        # IDENTITY (Static)
        self.ticker = ticker
        
        # STATE (Mutable & Persistent)
        self.history = deque(maxlen=window_size)
        self.ath = 0.0  # All-Time High
        self.session_count = 0
        self.start_time = datetime.now()

    def get_live_price(self):
        """Mock Generator for live data (Replace with API logic)"""
        price = 60000.0
        while True:
            price += random.uniform(-100, 105) # Slight upward bias
            yield round(price, 2)
            time.sleep(1)

    def process_price(self, price):
        """The 'Detective' Logic: Analyzing the data stream"""
        self.session_count += 1
        self.history.append(price)
        
        # Update All-Time High (Persistence)
        if price > self.ath:
            self.ath = price
            print(f"--- 🚀 NEW RECORD: {self.ticker} hit ${self.ath:,} ---")

        # Calculate Rolling Average (Efficiency)
        if len(self.history) == self.history.maxlen:
            sma = sum(self.history) / len(self.history)
            diff = ((price - sma) / sma) * 100
            
            print(f"[{datetime.now().strftime('%H:%M:%S')}] {self.ticker}: ${price:,} | "
                  f"SMA_{self.history.maxlen}: ${sma:,.2f} | Vol: {diff:+.4f}%")

    def run(self, limit=None):
        """The 'Engine': Orchestrating the flow"""
        print(f"Sentinel activated for {self.ticker}...")
        print(f"Starting balance/limit: ${self.ath}")
        
        price_stream = self.get_live_price()
        
        try:
            for current_price in price_stream:
                self.process_price(current_price)
                
                # Optional stop condition
                if limit and self.session_count >= limit:
                    break
        except KeyboardInterrupt:
            self.shutdown()

    def shutdown(self):
        duration = datetime.now() - self.start_time
        print(f"\n--- Sentinel Deactivated ---")
        print(f"Ticker: {self.ticker} | Session High: ${self.ath:,}")
        print(f"Total Ticks: {self.session_count} | Runtime: {duration}")

# --- EXECUTION ---
bot = MarketSentinel("BTC-USD", window_size=10)
bot.run(limit=50) # Run for 50 ticks and see the 'state' in action
```
