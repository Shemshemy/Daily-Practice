```python
from collections import deque

class LendingBot:
    def __init__(self, multiplier=5):
        # 1. Initialize a deque called 'deposits' with a maxlen of 3
        deposits = deque(maxlen=3)
        # 2. Store the 'multiplier' as a permanent attribute
        self.multiplier = multiplier
        pass

    def add_deposit(self, amount):
        # 1. Add the 'amount' to your deque
        self.deposits.append(amount)
        # 2. Print a confirmation message
        print(f'{self.amount} wass added to deposits')
        

    def get_limit(self):
        # 1. Check if the deque is empty (to avoid errors)
        if  not self.deposits:
            return 0
        # 2. Calculate the average of the items in the deque
        average_of_items = self.deposits / len(self.deposits)
        # 3. Multiply that average by your 'self.multiplier'
        limit =  average_of_items * self.multiplier
        # 4. Return the result
        return round (limit, 2)
```
