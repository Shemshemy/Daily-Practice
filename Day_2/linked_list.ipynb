How much faster are linked lists compared to normal lists? 

```python
import time
from collections import deque

# We'll use 100,000 items to make the computer sweat a little
N = 1_000_000

# --- RACE 1: The Standard List (Array) ---
test_list = list(range(N))
start_time = time.time()

# Inserting at the beginning 1,000 times
for _ in range(1000):
    test_list.insert(0, "NEW_DATA")

list_duration = time.time() - start_time
print(f"Standard List Time: {list_duration:.5f} seconds")


# --- RACE 2: The Deque (Linked List) ---
test_deque = deque(range(N))
start_time = time.time()

# Inserting at the beginning 1,000 times
for _ in range(1000):
    test_deque.appendleft("NEW_DATA")

deque_duration = time.time() - start_time
print(f"Deque (Linked List) Time: {deque_duration:.5f} seconds")

# --- THE VERDICT ---
multiplier = list_duration / deque_duration
print(f"\nResult: The Linked List logic was {multiplier:.1f}x faster!")
```
