```python
# Implementation of a house with colour and size attributes
class house:
    def __init__(self, colour, size):
        self.colour = colour
        self.size = size

    def describe(self):
        print(f"This house is {self.colour} and has {self.size:,} rooms")
```

```python
# A couple of instances of the class
mine = house("Red", 3)
brians = house("black", 5)
```

```python
# Describing them
mine.describe()
brians.describe()
```

```python
class wallet:
    def __init__(self, balance, withdraw):
        self.balance = balance
        self.withdraw = withdraw

    def calculation(self):
        state_of_withdrawal = self.balance - self.withdraw
        if state_of_withdrawal < 0:
            print("you have insufficient balance")
        else:
            print(f"{state_of_withdrawal}")

```

```python
kim = wallet(2000, 5000)
```

```python
kim.calculation()
```

```python
class users:
    def __init__(self, status):
        self.status = status

    def toogle_status(self):
        if self.status == "online":
            self.status = "offline"
        elif self.status == "offline":
            self.status = "online"
        print(f'{self.status}')

```

```python
Dennis = users("offline")
```

```python
Dennis.toogle_status()
```
