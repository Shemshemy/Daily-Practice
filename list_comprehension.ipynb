#### The Muscle Memory Exercises
1. The Warm-up (Basic Filtering)
You have a list of parcel sizes in acres: sizes = [0.5, 1.2, 25.0, 0.8, 15.4, 3.0].

Task: Use a list comprehension to create a new list containing only "Large Scale" parcels (anything over 10 acres).

```python
# implementation in code
sizes = [0.5, 1.2, 25.0, 0.8, 15.4, 3.0]
greater_than_10 = [size for size in sizes if size > 10]
greater_than_10
```

2. The Transformation (Lambda + Map)
You have a list of owner names in lowercase: owners = ["dennis", "shem", "limo", "rogers"].

Task: Use map() and a lambda function to capitalize every name in the list.

```python
owners = ["dennis", "shem", "limo", "rogers"]
capitalized = list(map(lambda owner: owner.capitalize(), owners))
capitalized 
```

3. The Conditional Shift (If-Else in Comprehension)
You have a list of land categories: land_use = ["Residential", "Agricultural", "Public", "Residential"].

Task: Create a list that replaces "Residential" with "Area_A" and anything else with "Area_B".

Hint: [x if condition else y for x in list]

```python
land_use = ["Residential", "Agricultural", "Public", "Residential"]
corrected = ['Area_A' if area == 'Residential' else 'Area_B' for area in land_use]
corrected
```

4. The Data Cleaner (Nested Logic)
You have a list of strings representing ID numbers, but some have accidental whitespace: ids = ["  K001", "K002  ", " K003 ", "K004"].

Task: Use a list comprehension to strip the whitespace and append the suffix _VALID to each ID (e.g., "K001_VALID").

```python
ids = ["  K001", "K002  ", " K003 ", "K004"]
stripped = [id.strip()+'_VALID' for id in ids]
stripped
```

5. The "Boss Level" (Dictionary Comprehension + Logic)
You have two related lists: parcels = ["A1", "A2", "A3"] and values_ksh = [500000, 1200000, 800000].

Task: Create a dictionary where the key is the parcel ID and the value is the tax amount (calculated as 5% of the value), but only if the tax is greater than 30,000. Use a lambda or a simple expression within the comprehension.

```python
parcels = ["A1", "A2", "A3"] 
values_ksh = [500000, 1200000, 800000]

taxes = [(parcel, 0.05 * value) for parcel, value in zip(parcels, values_ksh) if (value*.05)>30000]
print(taxes)
```
