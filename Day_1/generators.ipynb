1. The Basic Yield (Sequential Data)
Goal: Master the yield keyword to create a custom range of transaction IDs.

```python
# Task: Complete this function to yield "TX-1", "TX-2" up to "TX-n"
def generate_transactions(n):
    for i in range(1, n + 1):
        yield(f'TX-{i}')
        

# Validation:
# gen = generate_transactions(5)
# print(next(gen)) # Should output TX-1
gen = generate_transactions(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

2. The Memory Duel (List vs. Generator)
Goal: Prove the efficiency of a Generator Expression over a List Comprehension.

```python
import sys

# Task:
# 1. Create a list of 1,000,000 squares using []
# 2. Create a generator for 1,000,000 squares using ()
# 3. Compare their memory sizes using sys.getsizeof()

data_range = range(1000000)

# Implementation:
list_comp = [item for item in data_range]
def generator(data_range):
    for data in data_range:
        yield(data)

gen_exp = generator(data_range)
print(f"List Size: {sys.getsizeof(list_comp)} bytes")
print(f"Gen Size: {sys.getsizeof(gen_exp)} bytes")
```

3. The Data Sieve (Conditional Streaming)
Now, let's combine the List Comprehension skills from earlier with your new Generator skills.

The Context: You are filtering a stream of job postings. You want to see "Remote" roles without loading the entire database into memory.

```python
raw_records = [
    {'id': 101, 'type': 'Remote', 'salary': 120000},
    {'id': 102, 'type': 'On-Site', 'salary': 95000},
    {'id': 103, 'type': 'Remote', 'salary': 140000},
    {'id': 104, 'type': 'Hybrid', 'salary': 110000},
    {'id': 105, 'type': 'Remote', 'salary': 130000},
]

# Task: Write a generator function 'remote_filter' that 
# yields ONLY the IDs of 'Remote' jobs.

def remote_filter(records):
    for job in records:
        if job['type'] == "Remote":
            yield(job)


# Validation:
for job_id in remote_filter(raw_records):
    print(f"Processing Remote Job ID: {job_id}")
```

4. The Infinite Stream (Logic Control)
This tests my ability to manage loops that don't have a defined end—crucial for streaming real-time data.

```python
# Task: Create a generator that yields powers of 2 starting from 2 (2, 4, 8, 16...)
# Use a 'while True' loop inside.
def power_of_two():
    current_value = 2
    while True:
        yield current_value
        current_value *= 2

# Validation:
gen = power_of_two()

# Use a loop to print values, but 'break' once the value is > 2000.
for val in gen:
    if val > 2000:
        break
    print(val)
```
