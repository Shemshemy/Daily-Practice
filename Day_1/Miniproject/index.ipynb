```python
import yfinance as yf
import time
from datetime import datetime


def live_price_generator(ticker_symbol):
    ticker = yf.Ticker(ticker_symbol)
    
    while True:
        # Pull the most recent 1-minute interval data
        # 'period="1d"' gets today, 'interval="1m"' gets 1-minute bars
        data = ticker.history(period="1d", interval="1m")
        
        if not data.empty:
            # The very last row [-1], the 'Close' column
            current_price = data['Close'].iloc[-1]
            yield current_price
        
        # Keep the sleep at 5-10 seconds so Yahoo doesn't block my IP
        time.sleep(10)


# Implementation Block
def monitor_market():
    # 1. Initialize the generator
    prices = live_price_generator("BTC-USD")
    
    # 2. Grab the first price to set a baseline
    last_reported_price = next(prices)
    print(f"Monitoring started. Baseline Price: ${last_reported_price:,.2f}")
    
        
    # 3. The Monitor Loop
    for current_price in prices:
        difference = current_price - last_reported_price
        now = datetime.now()
        # Use abs() to catch both jumps and drops
        if abs(difference) >= .01:
            direction = "UP 📈" if difference > 0 else "DOWN 📉"
            log_entry = f"[{now.strftime('%Y-%m-%d %H:%M:%S')}] ALERT: {direction} move of ${abs(difference):.2f}. Price: ${current_price:,.2f}\n"            
            # Print it
            print(log_entry)
            
            # Save it (The "A" means Append - it won't delete old data)
            with open("market_alerts.txt", "a", encoding = 'utf-8') as f:
                f.write(log_entry)
                
            last_reported_price = current_price


if __name__ == "__main__":
    monitor_market()
```
