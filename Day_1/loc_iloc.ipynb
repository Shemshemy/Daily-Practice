```python
import pandas as pd

# Mock Data: A small server log
data = {
    'Level': ['INFO', 'ERROR', 'DEBUG', 'ERROR', 'INFO'],
    'Message': ['System Start', 'Database Timeout', 'Checking Port', 'Auth Failure', 'User Login'],
    'Latency': [10.5, 450.2, 5.1, 89.0, 12.3]
}

# We set the index to specific 'Log_IDs' to practice .loc
df = pd.DataFrame(data, index=['L101', 'L102', 'L103', 'L104', 'L105'])

# Display to verify
print(df)
```

Problem Set 1: The Navigator (loc & iloc)
Goal: Retrieve specific evidence from the logs without crashing the system.

1. The Target: Use .loc to find the Message for the log ID 'L104'.

2. The Scan: Use .iloc to get the last two rows of the entire DataFrame.

3. The Specifics: Use .iloc to find the Latency (which is column index 2) of the first row.

4. The Filter: (Advanced) Use .loc to find all rows where the Level is 'ERROR'.

Hint: df.loc[df['Level'] == 'ERROR']

```python
#The Target: Use .loc to find the Message for the log ID 'L104'.
df.loc['L104']['Message']
```

```python
#The Scan: Use .iloc to get the last two rows of the entire DataFrame.
df.iloc[-2:]
```

```python
# The Specifics: Use .iloc to find the Latency (which is column index 2) of the first row.
df.iloc[1]['Latency']
```

```python
# The Filter: (Advanced) Use .loc to find all rows where the Level is 'ERROR'.
df.loc[df['Level'] == 'ERROR']
```
